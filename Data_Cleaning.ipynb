```python
import pandas as pd
from google.colab import files

# 1️⃣ Load dataset
station_hour = pd.read_csv("/content/station_hour.csv")

# 2️⃣ Strip leading/trailing spaces from column names
station_hour.columns = station_hour.columns.str.strip()

# 3️⃣ Convert Datetime column to proper datetime
station_hour['Datetime'] = pd.to_datetime(station_hour['Datetime'], errors='coerce')

# 4️⃣ Sort by StationId and Datetime
station_hour = station_hour.sort_values(['StationId', 'Datetime'])

# 5️⃣ Identify numeric columns
numeric_cols = station_hour.select_dtypes(include=['float64', 'int64']).columns

# 6️⃣ Drop rows where all numeric columns are missing
station_hour = station_hour.dropna(subset=numeric_cols, how='all')

# 7️⃣ Fill numeric columns
# Step 1: Forward + backward fill per station
for col in numeric_cols:
    station_hour[col] = station_hour.groupby('StationId')[col].apply(lambda x: x.ffill().bfill()).values

# Step 2: Fill remaining NaNs with station-wise median
for col in numeric_cols:
    station_hour[col] = station_hour.groupby('StationId')[col].apply(lambda x: x.fillna(x.median())).values

# Step 3: Fill any remaining NaNs with overall median
for col in numeric_cols:
    station_hour[col].fillna(station_hour[col].median(), inplace=True)

# 8️⃣ Fill categorical column AQI_Bucket with mode
if 'AQI_Bucket' in station_hour.columns:
    station_hour['AQI_Bucket'] = station_hour['AQI_Bucket'].fillna(station_hour['AQI_Bucket'].mode()[0])

# 9️⃣ Format Datetime column for readability
station_hour['Datetime'] = station_hour['Datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')

# 🔟 Optional: Check missing values after cleaning
print("Missing values after cleaning:\n", station_hour.isna().sum())

# 1️⃣1️⃣ Save and download cleaned CSV
station_hour.to_csv("cleaned_station_hour.csv", index=False)
files.download("cleaned_station_hour.csv")
```

```python
import pandas as pd
from google.colab import files
import numpy as np

# Load dataset
real = pd.read_csv("/content/Real time Air Quality Index from various locations.csv")

# Strip column names
real.columns = real.columns.str.strip()

# Convert 'na' or 'Na' strings to np.nan
real.replace(['na', 'Na', 'NA'], np.nan, inplace=True)

# Numeric columns to fill
numeric_cols = ['pollutant_min', 'pollutant_max', 'pollutant_avg']

# Fill missing numeric values (station-wise median using transform)
for col in numeric_cols:
    real[col] = real.groupby('station')[col].transform(lambda x: x.fillna(x.median()))

# Fill any remaining NaNs with overall median
for col in numeric_cols:
    real[col].fillna(real[col].median(), inplace=True)

# Check missing values
print(real[numeric_cols].isna().sum())

# Save and download cleaned CSV
real.to_csv("cleaned_realtime_aqi.csv", index=False)
files.download("cleaned_realtime_aqi.csv")


```
